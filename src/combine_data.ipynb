## JobsDB data

```python
BA = pd.read_excel("../data/BA_25Processed.xlsx", index_col=0)
DA = pd.read_excel("../data/DA_25Processed.xlsx", index_col=0)
DE = pd.read_excel("../data/DE_25Processed.xlsx", index_col=0)
DS = pd.read_excel("../data/DS_25Processed.xlsx", index_col=0)
MLE = pd.read_excel("../data/MLE_25Processed.xlsx", index_col=0)
talent_connect = pd.read_excel("../data/TalentConnectProcessed.xlsx", index_col=0)
```

```python
combined_df = pd.concat([BA, DA, DE, DS, MLE])
combined_df
```

```python
cleaned_df = combined_df.drop_duplicates(subset=['job-title', 'company', 'location', 'date', 'Job description'], keep="first")
cleaned_df
```
